```python
import pandas as pd
import numpy as np
import matplotlib.colors as mcolors
```

```python
county_msas = pd.read_excel('Counties_MSAS.xlsx')
```

```python
claims = pd.read_excel('weekly-claims-by-county-twc.xlsx')
```

```python
counties = county_msas.dropna()
```

```python
claims.head()
```

```python
county_msas.loc[county_msas['Core Based Statistical Area'] == 'Dallas-Fort Worth-Arlington, TX MSA']
```

```python
county_msas['County'][100][0:-14]
```

```python
county_list = []
for index, value in counties['County'].items():
    county_list.append(value[0:-14])
```

```python
len(county_list) == len(counties['County'])
```

```python
counties['County'] = list(county_list)
```

```python
counties.head()
```

```python
claims_clean = claims[claims['County'].isin(county_list)].sort_values(by = 'County').reset_index(drop = True)
```

```python
counties_clean = counties.sort_values(by = 'County').reset_index(drop=True)
```

```python
counties_clean.head()
```

```python
claims_df = claims_clean.join(counties_clean, lsuffix = 'f')
```

```python
claims_df.head()
```

```python
claims_df.loc[claims_df['Core Based Statistical Area'].str.contains('MSA')]
```

```python
c = claims_df.loc[claims_df['Core Based Statistical Area'].str.contains('MSA')]
c.head()
```

```python
c = c.reset_index(drop = True)
c = c.fillna(0)
c.head()
```

```python
final_county_msa_df = c.drop(columns = ['Countyf','2019 Population.1'])
```

```python
msa_df = final_county_msa_df.drop(columns = 'County')
msa_df.head(5)
```

```python
final_df = msa_df.groupby(by = 'Core Based Statistical Area').sum().T
final_df.head(5)
# final_df.to_csv('dataframe.csv')
```

```python
import datetime
final_df.index[0]
```

```python
final_df = final_df.loc[datetime.datetime(2020, 7, 18, 0, 0): datetime.datetime(2020,3,7,0,0)]
```

```python
color_list = list(mcolors.CSS4_COLORS.keys())
colors = []
for i in range(27):
    colors.append(color_list[4*i])
```

```python
color_column = colors*20
```

```python
final_df.head()
```

```python
multi_df = final_df.stack()
multi_df
```

```python
multi_df.index
#data.loc[data['URL'].str.contains("youtub"),'test'] = 'Youtube' #example of assign value per msa
```

```python
multi_df = multi_df.rename_axis(['Dates', 'MSA'])
multi_df
```

```python
multi_df.index
```

```python
multi_frame = multi_df.to_frame()
multi_frame.columns = ['Data']
multi_frame.index[0][1]
```

```python
multi_frame['Dates'] = multi_frame.index.get_level_values(0)
```

```python
multi_frame['MSA'] = multi_frame.index.get_level_values(1)
```

```python
multi_frame['Dates']
```

```python
multi_frame['color_column'] = color_column
```

```python
multi_frame.tail()
```

```python
from bokeh.io import output_notebook, show
from bokeh.plotting import figure, output_file, save
from bokeh.models import ColumnDataSource
from bokeh.models.tools import HoverTool
import bokeh.models.formatters as formatthis

from bokeh.resources import CDN
from bokeh.embed import file_html
```

```python
output_notebook()
```

```python
dource = ColumnDataSource(multi_frame)
g = figure(x_axis_type="datetime",plot_height=600, plot_width=700, title="Weekly Unemployment Claims")
g.circle(x = 'Dates', y = 'Data', source = dource,size =8 ,color= 'color_column',line_color = 'black')
g.xgrid.grid_line_alpha = 1
g.xgrid.grid_line_color = 'black'
g.ygrid.grid_line_color = 'black'
hover = HoverTool()
hover.formatters = {'@Dates': "datetime"}
hover.tooltips=[
    ('Claims', "@Data{0,0}"),
    
    ('MSA','@MSA')
]

g.add_tools(hover)
#darker lines
#add more
show(g)
output_file('Weekly_claims_bokeh.html')
save(g)
```
